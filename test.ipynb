```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N_4 = 50
N_2 = N_4 * 2
N = N_2 * 2

mat = np.array([[0, 1, 0.5, 1] * N_4]).reshape(1, N)
# mat = np.array([[0, 1] * N_2, [1, 0] * N_2, [0, 1, 0, 1] * N_4, [1, 0] * N_2, [0, 1] * N_2, [1, 0] * N_2, [0, 1, 0, 1] * N_4, [1, 0] * N_2]).reshape(8, N)
plt.figure(figsize=(20, 20))
plt.imshow(mat)

fft = np.fft.fft2(mat)
print(np.argmax(np.abs(fft)))
plt.figure(figsize=(20, 20))
plt.imshow(np.abs(fft))
print(fft[0,0])
```
